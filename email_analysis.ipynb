```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import altair as alt
```

```python
email_data = "./usage_data.csv"
data = pd.read_csv(email_data)

data.head()
```

```python
data.describe()
```

```python
data.isnull().values.any()
```

```python
data["Date"] = pd.to_datetime(data["Date"], errors="coerce")
```

```python
data23 = data[data["Date"].dt.year == 2023]
data22 = data[data["Date"].dt.year == 2022]
data22_23 = data[data["Date"].dt.year == 2022 & 2023]
data23
```

```python
alt.Chart(data23).mark_circle(size=60).encode(
    x="Date",
    y="Average time spent",
    color="Feature",
    tooltip=["Username", "Sessions", "Feature", "Average time spent"],
).interactive()
```

```python
feature_count22 = data22['Feature'].value_counts().reset_index()
feature_count22
```

```python
feature_count23 = data23['Feature'].value_counts().reset_index()
feature_count23
```

```python
base22 = alt.Chart(data22).encode(x="Date")

bar22 = base22.mark_bar().encode(y="Time spent")

line22 = base22.mark_line(color="red").encode(y="Average time spent")

(bar22 + line22).properties(width=825)
```

```python
daily_count = data23.groupby("Date").size().reset_index(name="Count")
print(daily_count)
```

```python
username_count = data22["Username"].value_counts().reset_index()
username_count.columns = ["Username", "Count"]
```

```python
username_count
```

```python
pre_user_stats22 = (
    data22.groupby("Username")
    .agg({"Sessions": "sum", "Time spent": "sum", "Average time spent": "mean"})
    .assign(Count=lambda x: data22["Username"].value_counts())
    .reset_index()
)
user_stats22 = pre_user_stats22.rename(
    columns={
        "Username": "username",
        "Sessions": "sessions",
        "Time spent": "time_spent",
        "Average time spent": "avg_time_spent",
        "Count": "count",
    }
)
pre_user_stats23 = (
    data23.groupby("Username")
    .agg({"Sessions": "sum", "Time spent": "sum", "Average time spent": "mean"})
    .assign(Count=lambda x: data22["Username"].value_counts())
    .reset_index()
)
user_stats23 = pre_user_stats23.rename(
    columns={
        "Username": "username",
        "Sessions": "sessions",
        "Time spent": "time_spent",
        "Average time spent": "avg_time_spent",
        "Count": "count",
    }
)
user_stats23
```

```python
user_stats22
```

```python
test_chart23 = (
    alt.Chart(user_stats23)
    .mark_bar(opacity=0.7)
    .encode(
        x="username:N",  # 'N' for nominal data type
        y=alt.Y("sessions:Q", stack=None),  # 'Q' for quantitative, no stacking
        color="time_spent:Q",  # Color based on time spent
        tooltip=[
            "username:N",
            "sessions:Q",
            "time_spent:Q",
            "avg_time_spent:Q",
            "count:Q",
        ],
    )
)
test_chart23
```

```python
# Base chart for 2023 data
chart_22 = (
    alt.Chart(user_stats22)
    .mark_bar(opacity=0.7)
    .encode(
        x="username:N",
        y=alt.Y("sessions:Q", stack=None),
        color=alt.value("blue"),  # Fixed color for 2023 data
        tooltip=["username", "sessions", "time_spent", "avg_time_spent", "count"],
    )
)

# Base chart for 2024 data
chart_23 = (
    alt.Chart(user_stats23)
    .mark_bar(opacity=0.7)
    .encode(
        x="username:N",
        y=alt.Y("sessions:Q", stack=None),
        color=alt.value("red"),  # Fixed color for 2024 data
        tooltip=["username", "sessions", "time_spent", "avg_time_spent", "count"],
    )
)

layered_chart = chart_22 + chart_23
layered_chart
```
