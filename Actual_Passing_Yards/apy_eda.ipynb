```python
import pandas as pd
import numpy as np
from ydata_profiling import ProfileReport
from skimpy import skim
import altair as alt

df = pd.read_csv("apy.csv")
```

```python
df['apy'] = df['yards'] - df['yac']
df
```

```python
df['apy/g'] = np.ceil(df['apy'] / df['gp'] * 100)/100
df['tot_td'] = df['td'] + df['rtd']
df
```

```python
apy_g = df.sort_values(by=['apy/g','tot_td'], ascending=False)
apy_g
```

```python
profile = ProfileReport(df, title="Profiling Report")
profile
```

```python
alt.Chart(df).mark_circle(size=60).encode(
    x='td',
    y='apy/g',
    color='Name',
    tooltip=['Name', 'apy', 'td', 'rtd','tot_td','gp']
    ).interactive()
```
