```python
def get_region_time_spent(
    month: str, df_feature_usage_data: pd.DataFrame, df_user_data: pd.DataFrame
) -> pd.DataFrame:
    # Convert month string to datetime object and calculate the start and end dates for last year
    start_date = pd.to_datetime(f"{int(month[:4]) - 1}-08-01")
    end_date = pd.to_datetime(month)

    # Merge with user_data to get Region information
    merged_data = pd.merge(
        df_feature_usage_data, df_user_data, on="username", how="left"
    )
    merged_data["date"] = pd.to_datetime(merged_data["date"])
    filtered_data = merged_data[
        (merged_data["date"] >= start_date) & (merged_data["date"] <= end_date)
    ]
    # Group by Region and sum up time spent; handle 'NaN' regions as "No-Region"
    region_time_spent = merged_data.groupby("Region")["time spent"].sum().reset_index()
    region_time_spent.loc[region_time_spent["Region"].isna(), "Region"] = "No-Region"

    return region_time_spent
```
